## 2. Stopwords

```python
from nltk.corpus import stopwords
```

```python
example_text1 = "this is an example showing off stop word filtration"
```

```python
stop_words = set(stopwords.words("english"))
stop_words
```

```python
words = word_tokenize(example_text1)
```

```python
filtered_sentence = []
stopword_list = []
for w in words:
    if w not in stop_words:
        filtered_sentence.append(w)
    elif w in stop_words:
        stopword_list.append(w)
print(filtered_sentence)
stopword_list
```

```python
len(stop_words)
```

```python
len(words) == len(filtered_sentence)+len(stopword_list)
```

### 4. Speech Tagging

```python
import nltk
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
```

```python
train_text = state_union.raw("2005-GWBush.txt")
sample_text = state_union.raw("2006-GWBush.txt")
```

```python
custom_sent_tokenizer = PunktSentenceTokenizer(train_text)
tokenized = custom_sent_tokenizer.tokenize(sample_text)
```

```python
def process_content():
    try:
        for i in tokenized[:5]:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            print(tagged)

    except Exception as e:
        print(str(e))


process_content()
```

### 6. Named entity recognition

```python
import nltk
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer as pst

train_text = state_union.raw('2005-GWBush.txt')
test_text = state_union.raw('2006-GWBush.txt')

model = pst(train_text)
testing_model = model.tokenize(test_text)

def process_content():
    try:
        for i in testing_model[5:]:
            words = nltk.word_tokenize(testing_model)
            tagged = nltk.pos_tag(words)
            namedEnt = nltk.ne_chunk(tagged, binary=True)
            namedEnt.draw()
    except Exception as e:
        print(str(e))
        
process_content()
```

```python
synonyms = []
antonyms = []

for syn in wordnet.synsets("good"):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())

print(set(synonyms))
print(set(antonyms))
```
