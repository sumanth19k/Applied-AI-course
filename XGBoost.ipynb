```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree
import random
```

```python
np.random.seed(42)

X = np.random.rand(100, 1)-0.5
y = 2*X[:, 0]**2+0.05*np.random.randn(100)
```

```python
df = pd.DataFrame()
```

```python
df['X'] = X.reshape(100)
df['y'] = y
```

```python
df.head()
```

```python
plt.scatter(df['X'], df['y'])
```

```python
df['pred1'] = df['y'].mean()
```

```python
df['res1'] = df['y']-df['pred1']

df.head()
```

```python
# second model
```

```python
tree = DecisionTreeRegressor(max_leaf_nodes=8)

# best results between 8 and 32
```

```python
tree.fit(df['X'].values.reshape(100, 1), df['res1'].values)
```

```python
plt.figure(figsize=(12,12))  # set plot size (denoted in inches)
plot_tree(tree, fontsize=10)
plt.show()
```

```python
X_test = np.linspace(-0.5,0.5,100)

y_pred = 0.176954+(tree.predict(X_test.reshape(100, 1)))
```

```python
plt.plot(X_test, y_pred, linewidth=3)
plt.scatter(df['X'], df['y'], color='red')
```
